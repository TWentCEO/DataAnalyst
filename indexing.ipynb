```python
dict_prd = {
    'L': [5141, 5365, 5084],
    'M': [4968, 4977, 4915],
    'S': [5012, 4991, 5029],
    'XL': [5071,4976, 5023],
    'XS': [5195, 4954, 5088],
}
print(dict_prd)
print(type(dict_prd))
```
